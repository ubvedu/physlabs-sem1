```python
import numpy as np

l = np.array([5.6, 7.2, 9.6, 11.6, 13.6, 15.6])
T = np.array([
    [18.91, 18.96, 18.97, 19.00],
    [23.85, 23.98, 23.97, 24.00],
    [28.87, 28.86, 28.77, 28.86],
    [34.15, 33.90, 33.92, 33.97],
    [39.04, 39.11, 39.35, 39.30],
    [44.48, 44.75, 44.69, 44.53],
])
d = 1990 * 10 ** -6
T_self = np.array([11.33, 11.44, 11.43, 11.33])
```

```python
T_mean = np.array([a.mean() for a in T])
T_self_mean = T_self.mean()
n = len(T[0])
T_mean_err = [np.sqrt(sum((T[i] - T_mean[i]) ** 2) / (n * (n - 1))) for i in range(len(T))]
l_err = 0.05
```
